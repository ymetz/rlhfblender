```python
from parametric_angle_umap import ParametricAngleUMAP as pmap
import umap
import numpy as np
import random
```

```python
singel_step_len = 512
actions = np.reshape([[[1],[0],[3],[2]] for _ in range(singel_step_len)], (singel_step_len*4,1))
nr_of_actions = 4
angle_array = np.arange(0, 2 * np.pi, 2 * np.pi / nr_of_actions)
# Action is an index, create array of actions, ignore last action as we can only compute
# the angle loss for the first n-1 actions
actions_angles = np.array([angle_array[action] for action in actions[:-1]])
```

```python
spread = 0.1
obs = np.reshape([[[0+random.uniform(-spread,spread),0+random.uniform(-spread,spread)],[0+random.uniform(-spread,spread),1+random.uniform(-spread,spread)],[1+random.uniform(-spread,spread),1+random.uniform(-spread,spread)],[1+random.uniform(-spread,spread),0+random.uniform(-spread,spread)]] for _ in range(singel_step_len)], (singel_step_len*4,2))
```

```python
angle_umap = pmap(angle_loss_weight=0.0)
mapper = angle_umap.fit(obs, action_angles=actions_angles)
```
